# CIFRADO POR SUSTITUCIÃ“N (Cifrado CÃ©sar)

Este programa implementa un cifrado clÃ¡sico de **SUSTITUCIÃ“N**, tambiÃ©n conocido como **CIFRADO CÃ‰SAR**.

A diferencia del cifrado por permutaciÃ³n (donde las letras cambian de posiciÃ³n), en el cifrado por sustituciÃ³n las **POSICIONES permanecen iguales**, pero cada **LETRA se reemplaza por otra** siguiendo un **DESPLAZAMIENTO numÃ©rico** en el alfabeto.

## Ejemplo simple:

- **Texto original:** `HELLO`
- **Desplazamiento:** `3`
- **Proceso:**
  - `H` â†’ desplaza 3 posiciones â†’ `K`
  - `E` â†’ desplaza 3 posiciones â†’ `H`
  - `L` â†’ desplaza 3 posiciones â†’ `O`
  - `L` â†’ desplaza 3 posiciones â†’ `O`
  - `O` â†’ desplaza 3 posiciones â†’ `R`
- **Resultado cifrado:** `KHOOR`

## Â¿CÃ³mo funciona?

Cada letra se convierte a su valor ASCII, se le suma el desplazamiento, y usando el operador mÃ³dulo (`% 26`) se asegura que el resultado permanezca dentro del alfabeto. El proceso inverso (restando el desplazamiento) permite recuperar el texto original.

### CaracterÃ­sticas:

- ðŸ”¸ Mantiene mayÃºsculas y minÃºsculas por separado
- ðŸ”¸ No modifica caracteres especiales (espacios, puntuaciÃ³n, nÃºmeros)
- ðŸ”¸ Es simÃ©trico: el mismo desplazamiento sirve para cifrar y descifrar (negativo)
- ðŸ”¸ Su seguridad es baja: solo hay 25 claves posibles

```python
def encrypt_message(texto, pos): #texto para desifrar, posiciÃ³n de remplazo
    encrypted_text = ""
    for char in texto:
        if 'a' <= char <= 'z': #verifica si es letra minuscula y obtiene el cÃ³digo ASCII de a como referencia
            inicio = ord('a')
            shifted_char = chr((ord(char) - inicio + pos) % 26 + inicio) #transforma el caracter a ASCII apolica el desplazamiento y lo devuelve como char
        elif 'A' <= char <= 'Z':
            inicio = ord('A')
            shifted_char = chr((ord(char) - inicio + pos) % 26 + inicio)
        else:
            shifted_char = char
        encrypted_text += shifted_char #se va aÃ±adiendo los carÃ¡cteres cifrados al texto encriptado
    return encrypted_text

def decrypt_message(texto, pos):
    return encrypt_message(texto, -pos)
```

```python
# Ejemplo
message = "Hello, world!"
varlor_encrip = 3
encrypted_message = encrypt_message(message, varlor_encrip)
print("Mensaje original:", message)
print("Valor de desplazamiento:", varlor_encrip)
print("Mensaje cifrado:", encrypted_message)

decrypted_message = decrypt_message(encrypted_message, varlor_encrip)
print("Mensaje descifrado:", decrypted_message)
```
